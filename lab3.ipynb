# THEORY

The symbol π (pi) is a mathematical constant that represents the ratio of the circumference of a circle to its diameter. No matter how large or small a circle is, this ratio always remains the same. The value of π is approximately 3.14159265, but since it is an irrational number, it cannot be written as a simple fraction. Its decimal digits continue forever without repeating any pattern, making it a non-terminating and non-recurring decimal.

The concept of π has been known since ancient times. The Babylonians and Egyptians used an approximate value of 3.125 in architecture and astronomy. Later, the Greek mathematician Archimedes developed the polygon method to obtain a much more accurate approximation of π. Over time, many mathematical methods, such as the Leibniz formula, Wallis product, and Monte Carlo method, have been developed to calculate π with greater accuracy.

In this experiment, the Monte Carlo simulation method is used to estimate the value of π. This method is based on random sampling. The Law of Large Numbers states that as the number of random samples increases, the average of the results approaches the expected value. Therefore, using more random points gives a more accurate estimate of π.
